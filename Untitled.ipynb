```python
import googletrans
```

```python
translator = googletrans.Translator(service_urls=['translate.googleapis.com'])
```

```python
with open('a1_lang_search_combined') as f:
    for line in f:
        
        line = line.replace('(language)','')
        line = translator.translate(line,dest ='te').text
        line = line.replace(']' , '](language)')
        print(line)
```
